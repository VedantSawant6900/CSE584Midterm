```python
def read_file_to_list(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    return [line.strip() for line in lines]

dataset_list = read_file_to_list("full.txt")
```

```python
dataset_list
```

```python
allxi=[]
for i in dataset_list:
    sentence = i[:len(i)//2]
    sentence_list=sentence.split(" ")[:-1]
    if len(sentence_list)>7:
        sentence_list=sentence_list[:7]
    final_sentence=" ".join(sentence_list)
    allxi.append(final_sentence)
#print(allxi)
xi = list(set(allxi))
#print(xi)
print(len(xi))
for i in xi:
    print(i)
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.2-1B")
model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-3.2-1B")
```

```python
# Check if pad token is set; if not, set it to eos_token
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# Resize model embeddings to account for any added tokens
model.resize_token_embeddings(len(tokenizer))
# Define the prompt 19 28
with open("xi.txt","r") as f:
    allprompt = f.readlines()
    print(allprompt)
# f = open("/content/drive/MyDrive/ML_Midterm/xijbloom.txt", "a+")
# for prompt in allprompt:
#     print(prompt)
l=len(allprompt)
with open("xijlama.txt", "a+") as f:
  for i,prompt in enumerate(allprompt):
    if 1:
      try:
          prompt = prompt.replace('\n','')
          # prompt = allprompt[4].replace('\n','')
          print(i,"/",l," Starting: ",prompt)
          # Tokenize the input
          inputs = tokenizer(prompt, return_tensors="pt", padding=True)

          # Generate text completion
          output_sequences = model.generate(
    inputs["input_ids"],
    attention_mask=inputs["attention_mask"],
    max_length=20,
    num_return_sequences=1,
    temperature=0.7,
    top_k=50,
    top_p=0.9,
    no_repeat_ngram_size=2,
    early_stopping=True,
    pad_token_id=tokenizer.eos_token_id
)


          # Decode and print the output
          full_sentence=""
          for generated_sequence in output_sequences:
              decoded_sequence = tokenizer.decode(generated_sequence, skip_special_tokens=True)
              full_sentence+=decoded_sequence
          full_sentence = full_sentence.replace("\n","")
          xij = full_sentence.split(".")[0]+"."+"\n"
          print(xij)
          f.write(xij)
      except Exception as e:
          print("Exception raised ",e)



# # Check if pad token is set; if not, set it to eos_token
# if tokenizer.pad_token is None:
#     tokenizer.pad_token = tokenizer.eos_token

# # Resize model embeddings to account for any added tokens
# model.resize_token_embeddings(len(tokenizer))

# # Define the prompt
# prompt = "Complete sentence: Last night I went"

# # Tokenize the input and create attention mask
# inputs = tokenizer(prompt, return_tensors="pt", padding=True)

# # Generate text completion with attention mask
# output_sequences = model.generate(
#     inputs["input_ids"],
#     attention_mask=inputs["attention_mask"],
#     max_length=20,
#     num_return_sequences=1,
#     temperature=0.7,
#     top_k=50,
#     top_p=0.9,
#     no_repeat_ngram_size=2,
#     early_stopping=True,
#     pad_token_id=tokenizer.pad_token_id  # Use pad_token_id set above
# )

# # Decode and print only the first sentence of the output
# for generated_sequence in output_sequences:
#     decoded_sequence = tokenizer.decode(generated_sequence, skip_special_tokens=True)
#     first_sentence = decoded_sequence.split('.')[0] + '.'
#     print(first_sentence)
```

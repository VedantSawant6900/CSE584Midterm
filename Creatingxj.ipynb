```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bigscience/bloom-560m")
model = AutoModelForCausalLM.from_pretrained("bigscience/bloom-560m")
#894 line number 1st iteration

```

```python
# Define the prompt
with open("newxi.txt","r") as f:
    allprompt = f.readlines()
    print(allprompt)
f = open("bloomxij.txt", "a+")
# for prompt in allprompt:
#     print(prompt)
l=len(allprompt)
for i,prompt in enumerate(allprompt):
    try:
        prompt = prompt.replace('\n','')
        # prompt = allprompt[4].replace('\n','')
        print(i,"/",l," Starting: ",prompt)
        # Tokenize the input
        inputs = tokenizer(prompt, return_tensors="pt")


        # Generate text completion
        output_sequences = model.generate(
            inputs["input_ids"],
            max_length=50,
            num_return_sequences=1,
            no_repeat_ngram_size=2,
            early_stopping=True,
            temperature=0.7,
            top_k=50,
            top_p=0.9
        )

        # Decode and print the output
        full_sentence=""
        for generated_sequence in output_sequences:
            decoded_sequence = tokenizer.decode(generated_sequence, skip_special_tokens=True)
            full_sentence+=decoded_sequence
        xij = full_sentence.split(".")[0]+"."+"\n"
        print(xij)
        f.write(xij)
    except Exception as e:
        print("Exception raised")
        
f.close()

```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("facebook/opt-1.3b")
model = AutoModelForCausalLM.from_pretrained("facebook/opt-1.3b")
```

```python
with open("/content/drive/MyDrive/mlmidterm/newxi.txt","r") as f:
    allprompt = f.readlines()
    print(allprompt)
# f = open("/content/drive/MyDrive/ML_Midterm/xijbloom.txt", "a+")
# for prompt in allprompt:
#     print(prompt)
l=len(allprompt)
with open("/content/drive/MyDrive/mlmidterm/xijoptpart2.txt", "a+") as f:
  for i,prompt in enumerate(allprompt):
    if i>=2516:
      try:
          prompt = prompt.replace('\n','')
          # prompt = allprompt[4].replace('\n','')
          print(i,"/",l," Starting: ",prompt)
          inputs = tokenizer(prompt, return_tensors="pt")


          output_sequences = model.generate(
    inputs["input_ids"],
    max_length=20,
    num_return_sequences=1,
    temperature=0.7,
    top_k=50,
    top_p=0.9,
    no_repeat_ngram_size=2,
    early_stopping=True
)


          full_sentence=""
          for generated_sequence in output_sequences:
              decoded_sequence = tokenizer.decode(generated_sequence, skip_special_tokens=True)
              full_sentence+=decoded_sequence
          full_sentence = full_sentence.replace("\n","")
          xij = full_sentence.split(".")[0]+"."+"\n"
          print(xij)
          f.write(xij)
      except Exception as e:
          print("Exception raised")


```
